La cuenta que m√°s menciona a la escuela, obviando la propia escuela es la de mujeresaseguir

```python
df.groupby('username')[['tweet']].count().sort_values('tweet',ascending=False)
```

Pasamos a datetime la columna date

```python
df['date']= pd.to_datetime(df['date'])
```

Creamos una nueva columna llamada month

```python
df['month'] = list(map(lambda x: x.month, df['date']))
```

En Abril es el mes que m√°s se menciona

```python
df.groupby('month')[['tweet']].count().sort_values('tweet',ascending=False)
```
